```python
import re, string, unicodedata
import nltk
from nltk import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
```

```python
#preprocessing
def removeStopword(str):
    stop_words = set(stopwords.words('english')) 
    word_tokens = word_tokenize(str) 
    filtered_sentence = [w for w in word_tokens if not w in stop_words]  
    filtered_sentence = [] 
    
    for w in word_tokens: 
        if w not in stop_words: 
            filtered_sentence.append(w)
    return ' '.join(filtered_sentence) #dijoin karna bentukannya list

#remove sentence which contains only one word
def removeSentence(str): 
    word = str.split()
    wordCount = len(word)
    if(wordCount<=1):
        str = ''
    
    return str


def cleaning(str):
#     #remove non-ascii
#     str = unicodedata.normalize('NFKD', str).encode('ascii', 'ignore').decode('utf-8', 'ignore')
#     #remove URLs
#     str = re.sub(r'(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:\'".,<>?«»“”‘’]))', '', str)
#     #remove punctuations
#     str = re.sub(r'[^\w]|_',' ',str)
#     #remove digit from string
#     str = re.sub("\S*\d\S*", "", str).strip()
#     #remove digit or numbers
#     str = re.sub(r"\b\d+\b", " ", str)
#     #to lowercase
#     str = str.lower()
#     #Remove additional white spaces
#     str = re.sub('[\s]+', ' ', str)
    str = re.sub(r"http\S+", "", str)
    str = re.sub(r'@[A-Za-z0-9]+','',str)
    str = re.sub('https?://[A-Za-z0-9./]+','',str)
    str = re.sub("[^a-zA-Z]", " ", str)
    str = re.sub(r"\b\d+\b", " ", str)#renove digit number
    str = re.sub(r'\W*\b\w{1}\b', '',str)
    str = str.lower()
       
    return str

def preprocessing(str):
    str = removeSentence(str)
    str = cleaning(str)
    str = removeStopword(str)
     
    
    return str
```

```python
#do preprocessing
import pandas as pd
import xlsxwriter

fo = pd.read_excel('dataset.xlsx', sheet_name='dataset') #read excel file
txt = fo['news']
workbook = xlsxwriter.Workbook('cleandata.xlsx')
worksheet = workbook.add_worksheet()

row = 0
col = 0

rowHeaders = ['text']
worksheet.write_row(row, col,  tuple(rowHeaders))
        
for t in txt:
    new_txt = preprocessing(t)
    rowValues = [new_txt]
    row += 1
    worksheet.write_row(row, col, tuple(rowValues))
    
workbook.close()
```

```python
fo = pd.ExcelFile('cleandata.xlsx')
df = pd.read_excel(fo, 'Sheet1')

text = df['text']
text_list =  [i.split() for i in text]
print(len(text_list))
```

```python
#Create Biagram & Trigram Models 
from gensim.models import Phrases
# Add bigrams and trigrams to docs,minimum count 10 means only that appear 10 times or more.
bigram = Phrases(text_list, min_count=10)
trigram = Phrases(bigram[text_list], min_count=10)

for idx in range(len(text_list)):
    for token in bigram[text_list[idx]]:
        if '_' in token:
            # Token is a bigram, add to document.
            text_list[idx].append(token)
    for token in trigram[text_list[idx]]:
        if '_' in token:
            # Token is a bigram, add to document.
            text_list[idx].append(token)
```

```python
from gensim import corpora, models
# Create a dictionary representation of the documents.
dictionary = corpora.Dictionary(text_list)
dictionary.filter_extremes(no_below=5, no_above=0.2)# no_below kemunculan ga boleh dibawah 5, 
print(dictionary)
```

```python
#build corpus
doc_term_matrix = [dictionary.doc2bow(doc) for doc in text_list]

print(len(doc_term_matrix))
print(doc_term_matrix[29])

tfidf = models.TfidfModel(doc_term_matrix)
corpus_tfidf = tfidf[doc_term_matrix]
```

```python
from gensim.models.coherencemodel import CoherenceModel
from gensim.models.lsimodel  import LsiModel
from gensim.corpora.dictionary import Dictionary
from numpy import array
#function to compute coherence values
def compute_coherence_values(dictionary, corpus, texts, limit, start, step):
    coherence_values = []
    model_list = []
    for num_topics in range(start, limit, step):
        model = LsiModel(corpus=corpus, id2word=dictionary, num_topics=num_topics)#membangun model dengan LSI
        model_list.append(model)
        coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')#menggunakan coherence tipe cv
        coherence_values.append(coherencemodel.get_coherence())
        
    return model_list, coherence_values
```

```python
start=2
limit=20
step=2
model_list, coherence_values = compute_coherence_values(dictionary, corpus=corpus_tfidf, 
                                                        texts=text_list, start=start, limit=limit, step=step)
#show graphs LSI
import matplotlib.pyplot as plt
x = range(start, limit, step)
plt.plot(x, coherence_values)
plt.xlabel("Num Topics")
plt.ylabel("Coherence score")
plt.legend(("coherence_values"), loc='best')
plt.show()
```

```python
# Print the coherence scores
#LSI
for m, cv in zip(x, coherence_values):
    print("Num Topics =", m, " has Coherence Value of", round(cv, 4))#semakin nilai coherence nya tinggi , banyak topik semakin baik, semakin sesuai jumlah topiknya
```

```python
from pprint import pprint

model = LsiModel(corpus=corpus_tfidf, id2word=dictionary, num_topics=8)#kenapa 8 karena nilai cohorence nya paling tinggi
pprint(model.print_topics())
```

```python
for idx, topic in model.print_topics(-1):
    print('Topic: {} Word: {}'.format(idx, topic))
```

```python
lsitopics = model.show_topics(formatted=False)
```

```python
from gensim.models.coherencemodel import CoherenceModel
from gensim.models import LdaModel
from gensim.corpora.dictionary import Dictionary
from numpy import array
#function to compute coherence values
def compute_coherence_values_lda(dictionary, corpus, texts, limit, start, step):
    coherence_values_lda = []
    model_list_lda = []
    for num_topics in range(start, limit, step):
        ldamodel = LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics, iterations=100)#membangun model dengan LSa
        model_list_lda.append(ldamodel)
    
        coherencemodel_lda = CoherenceModel(model=ldamodel, texts=texts, dictionary=dictionary, coherence='c_v')#menggunakan coherence tipe cv
        coherence_values_lda.append(coherencemodel_lda.get_coherence())
        
    return model_list_lda, coherence_values_lda
```

```python
start=2
limit=20
step=2
model_list_lda, coherence_values_lda = compute_coherence_values_lda(dictionary, corpus=corpus_tfidf, 
                                                        texts=text_list, start=start, limit=limit, step=step)
#show graphs
import matplotlib.pyplot as plt
x = range(start, limit, step)
plt.plot(x, coherence_values_lda)
plt.xlabel("Num Topics")
plt.ylabel("Coherence score")
plt.legend(("coherence_values"), loc='best')
plt.show()
```

```python
# Print the coherence scores LDA
for l, cv in zip(x, coherence_values_lda):
    print("Num Topics =", l, " has Coherence Value of", round(cv, 4))
```

```python
from pprint import pprint

ldamodel = LdaModel(corpus=corpus_tfidf, id2word=dictionary, num_topics=6)#kenapa 6 karena nilai cohorence nya paling tinggi
pprint(ldamodel.print_topics())
```

```python
for idx, topic in ldamodel.print_topics(-1):
    print('Topic: {} Word: {}'.format(idx, topic))
```

```python
ldatopics = ldamodel.show_topics(formatted=False)
```

```python
lsitopics = [[word for word, prob in topic] for topicid, topic in lsitopics]
ldatopics = [[word for word, prob in topic] for topicid, topic in ldatopics]
```

```python
lsi_coherence = CoherenceModel(topics=lsitopics, texts=text_list, dictionary=dictionary, window_size=10).get_coherence()
#nilai [:10] dihilangin
lda_coherence = CoherenceModel(topics=ldatopics, texts=text_list, dictionary=dictionary, window_size=10).get_coherence()

print("Nilai Coherence LSI : "+str(lsi_coherence))
print("Nilai Coherence LDA : "+str(lda_coherence))
print("Different : "+str(lda_coherence-lsi_coherence))
```

```python
import matplotlib.pyplot as plt
import numpy as np
def evaluate_bar_graph(coherences, indices):
    
    assert len(coherences) == len(indices)
    n = len(coherences)
    x = np.arange(n)
    plt.bar(x, coherences, width=0.2, tick_label=indices, align='center', alpha=0.8)
    plt.xlabel('Models')
    plt.ylabel('Coherence Value')
```

```python
evaluate_bar_graph([lsi_coherence, lda_coherence],
                   ['LSI', 'LDA'])
```
